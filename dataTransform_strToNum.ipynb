## 컬럼별 데이터 타입 동일화 : 함수 → 실행

```python
# 문자열 → 숫자 : 만원, 억원
# 돈
def parse_money2num(money_str) :
  # 필터링
  cle_str = ''.join(filter(lambda x: x.isdigit() or x =='.' or x == '-', money_str ))
  ##
  if money_str == '-' :
    return 0
  elif '만원' in money_str :
    return int(float(cle_str) * 1)
  elif '억원' in money_str :
    return int(float(cle_str) * 10000)
  elif '조원' in money_str :
    return int(float(cle_str) * 100000000)
  elif '경원' in money_str :
    return int(float(cle_str) * 1000000000000)
  else:
    return int(float(cle_str))

def parse_invest2num(invest_str) :
  cle_str = ''.join(filter(lambda x:x.isdigit() or x == '.' or x == '-', invest_str))
  if invest_str == '-' :
    return 0
  if invest_str == '비공개' :
    return 0
  if '만' in invest_str :
    return int(float(cle_str) * 1)
  elif '억' in invest_str :
    return int(float(cle_str) * 10000)
  else :
    return int(float(cle_str))


##
def money2int(col) :
  df[col + '(만원)'] = df[col].apply(parse_money2num)
  df.drop(col, axis=1, inplace=True)

##
def invest2int(col) :
  df[col + '(만원)'] = df[col].apply(parse_invest2num)
  df.drop(col, axis=1, inplace=True)

# 인원수
def parse_employees2num(employees_str) :
  cle_str = ''.join(filter(lambda x:x.isdigit() or x == '.', employees_str))
  return int(float(cle_str))

##
def people2int(col) :
  df[col] = df[col].apply(parse_employees2num)
```

```python
money_l = ['자본금', '누적투자유치금액', '연매출', '매출액_2020', '매출액_2021','영업이익_2020','영업이익_2021', '순이익_2020', '순이익_2021', '자산_2020', '자산_2021', '자본_2020', '자본_2021', '2022기준 시장규모']
money_l2 = ['최근투자유치']
people_l =['고용인원', '연간 퇴사율']

# 27개의 컬럼 들 중 dtype 변환 필요한 17개(14,1,2)컬럼만 담음
col_all = money_l + money_l2 + people_l

for col in col_all :
  if col in money_l :
    money2int(col)
  elif col in money_l2 :
    invest2int(col)
  else :
    people2int(col)
```

## 주소 형식 동일화

```python
## 주소변환
def add2add(col) :
  '''
  주소 변환
  '''
  df[col] = df[col].apply(unifi_add)

# '시', '특별시', '광역시' 등의 문자 제외

def unifi_add(add_str) :
  '''
  주소 통합
  '''
  cle_l = add_str.split()
  if cle_l[0] == '서울시' :
    cle_l[0] = '서울'
  elif cle_l[0] == '서울특별시' :
    cle_l[0] = '서울'
  elif cle_l[0] == '대전광역시' :
    cle_l[0] = '대전'
  elif cle_l[0] == '제주특별자치도' :
    cle_l[0] = '제주'
  elif cle_l[0] == '부산광역시' :
    cle_l[0] = '부산'
  elif cle_l[0] == '강원특별자치도' :
    cle_l[0] = '강원'
  elif cle_l[0] == '인천광역시' :
    cle_l[0] = '인천'

  #cle_str = ' '.join(cle_l[0],cle_l[1])
  # cle_str = f"{cle_l[0]} {cle_l[1]}"
  if len(cle_l) > 1 :
    cle_str = ' '.join([cle_l[0], cle_l[1]])
  else :
    cle_str = cle_l[0]


  # print(cle_str,)
  return cle_str

```

## 파생변수"년차" 생성 : 2023년 기준, 설립일자 사용

```python
from datetime import datetime

# 문자열 설립일자 to 날짜 객체형태
def parse_foundDate(date_str) :
  return datetime.strptime(date_str, '%Y-%m-%d')

# 년차를 구할거임
def cal_years_gap(foundDate) :
  today = datetime.today()
  diff = today.year - foundDate.year
  return diff

# 설립일자 to 날짜형식
df['설립일자'] = df['설립일자'].apply(parse_foundDate)

# 년차 구함
df['년차'] = df['설립일자'].apply(cal_years_gap)

df.drop('설립일자', axis=1, inplace=True)
```

## 파생변수 "자본잠식률" 생성 : (자본금 - 2021자본)/자본금

```python
df['자본잠식률'] = ((df['자본금(만원)'].astype(float) - df['자본_2021(만원)'].astype(float)) / df['자본금(만원)'].astype(float)) * 100 #.astype(float)
```

### 파생변수 "자본잠식률" 소숫점자리 동일화

```python
df['자본잠식률'] = df['자본잠식률'].round(2)
```
