## '주소'컬럼을 조건문의 인자로 적용시킬 때 나올 수 있는 데이터 형태에 대한 경우의 수를 고려한 함수 생성 실행

```python
def cal_score(address) :
  for keywords, score in weight.items() :
    if isinstance(keywords, str) :
    # 생성한 dict(weight)의 키워드가 하나인 경우 (= 키워드가 문자열인 경우)
      if keywords in address :
        return score
    else :
    # 생성한 dict(weight)의 키워드가 두 개 이상인 경우
      if any(keywords in address for keywords in keywords) :
        return score
  return 0
```

```python
df['주소 점수'] = df['주소'].apply(cal_score)
```
